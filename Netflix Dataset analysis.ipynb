```python
#List of categorical (text) columns where we want to fill missing values
categorical_columns = ['Genre', 'Tags', 'Languages', 'Country Availability',
                       'Director', 'Writer', 'Actors', 'View Rating', 'Production House']

# Fill missing values in categorical columns with 'Unknown'
netflix_data[categorical_columns] = netflix_data[categorical_columns].fillna('Unknown')

# Define numeric_columns - This was missing and caused the error
numeric_columns = ['Release Date', 'Runtime', 'IMDb Score', 'Rotten Tomatoes Score', 'Awards Received', 'Awards Nominated For','Boxoffice'] # Add any other relevant numeric columns

# Now you can create the DataFrame
pd.DataFrame(netflix_data[numeric_columns])
```

```python
# List of numeric columns where we want to fill missing values
numeric_columns = ['Hidden Gem Score', 'IMDb Score', 'Rotten Tomatoes Score',
                   'Metacritic Score', 'Awards Received', 'Awards Nominated For',
                   'Boxoffice', 'IMDb Votes']

# Convert 'Boxoffice' column to numeric before calculating the median
netflix_data['Boxoffice'] = pd.to_numeric(netflix_data['Boxoffice'].str.replace(',', '').str.replace('$', ''), errors='coerce')

# Fill missing values in numeric columns with the median value of each column
netflix_data[numeric_columns] = netflix_data[numeric_columns].fillna(netflix_data[numeric_columns].median())
pd.DataFrame(netflix_data[numeric_columns])
```

```python
#QUESTION 10 :Top countries analysis?
# Splitting the 'Country Availability' column by multiple countries (in case content is available in multiple regions)
netflix_data['Country List'] = netflix_data['Country Availability'].str.split(',')

# Exploding the list into multiple rows (so each country has its own row)
country_data = netflix_data.explode('Country List')

# Counting the number of releases per country
top_countries = country_data['Country List'].value_counts().head(10)

# Displaying the top 10 countries with the most Netflix releases
pd.DataFrame(top_countries).reset_index().rename(columns={'index': 'Country', 'Country List': 'Number of Releases'})
```

The code splits the 'Country Availability' column into multiple countries for content available in more than one region. It then uses the .explode() method to create separate rows for each country, allowing the dataset to reflect individual country availability. The .value_counts() method counts the number of releases per country, and the top 10 countries with the most releases are displayed in a DataFrame. This helps identify which countries have the most Netflix content available.

# **DATA ANALYSIS WITH matplotlib.pyplot**

```python
#Question 1: What is the distribution of IMDb scores?

import matplotlib.pyplot as plt

# Plotting the distribution of IMDb Scores
plt.hist(netflix_data['IMDb Score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of IMDb Scores')
plt.xlabel('IMDb Score')
plt.ylabel('Frequency')
plt.show()
```

 The code generates a histogram to visualize the distribution of IMDb scores in the dataset. It divides the scores into 20 bins, uses sky blue bars with black edges for aesthetics, and includes a title and axis labels for clarity. The plot helps identify patterns in the scores, such as their range, frequency, and distribution shape (e.g., normal, skewed, or uniform).

```python
#Question 2: How many movies vs series are there?

# Plotting the number of Movies vs Series
movie_vs_series = netflix_data['Series or Movie'].value_counts()
plt.bar(movie_vs_series.index, movie_vs_series.values, color=['blue', 'orange'])
plt.title('Number of Movies vs Series')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

The code creates a bar chart to compare the number of movies and series in the dataset. It calculates the counts of each type using value_counts() and visualizes them with blue and orange bars. The chart includes a title and labeled axes, making it easy to identify which type (movies or series) has a higher count in the dataset.
Why bar chart :  compare categorical data


```python
#Question 3: What is the trend of Netflix releases over the years?

# Convert 'Netflix Release Date' column to datetime objects
netflix_data['Netflix Release Date'] = pd.to_datetime(netflix_data['Netflix Release Date'])

# Plotting the trend of Netflix releases
netflix_data['Year'] = netflix_data['Netflix Release Date'].dt.year
release_trend = netflix_data['Year'].value_counts().sort_index()

plt.plot(release_trend.index, release_trend.values, marker='o', color='purple')
plt.title('Trend of Netflix Releases Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.grid(True)
plt.show()
```

 The code analyzes the trend of Netflix releases over the years by converting the 'Netflix Release Date' column to datetime format and extracting the release years. It then calculates the number of releases per year, sorts the data chronologically, and visualizes it using a line plot. The plot highlights changes in Netflix's release frequency over time, showing whether the number of releases has increased, decreased, or fluctuated across years.

```python
#QUESTION 4 : Compare Netflix releases yearly .
# Extract the year from the Netflix release date
netflix_data['Year'] = netflix_data['Netflix Release Date'].dt.year

# Count the number of Netflix releases per year
yearly_releases = netflix_data['Year'].value_counts().sort_index()

# Plotting the number of Netflix releases per year using a bar graph
plt.figure(figsize=(10,6))
plt.bar(yearly_releases.index, yearly_releases.values, color='red')

plt.title('Number of Netflix Releases by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Releases', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

The code compares Netflix releases by year using a bar chart. It extracts the release year from the 'Netflix Release Date' column, counts the number of releases per year, and plots the data. The bar chart, styled with red bars, provides a clear visual representation of how the volume of Netflix releases has varied annually. It includes a grid, rotated year labels, and axis labels for readability and better understanding of trends in Netflix's yearly content output.
Why bar chart : effectively displays and compares the number of Netflix releases for each year, which is categorical data


# **DATA ANALYSIS With Using Matplotlib and Seaborn**

```python
#Question 1: What is the relationship between IMDb score and Box Office earnings?

import seaborn as sns

# Plotting a scatter plot for IMDb Score vs Box Office
plt.figure(figsize=(8,6))
sns.scatterplot(x='IMDb Score', y='Boxoffice', data=netflix_data, color='blue')
plt.title('IMDb Score vs Box Office Earnings')
plt.xlabel('IMDb Score')
plt.ylabel('Box Office Earnings')
plt.grid(True)
plt.show()
```

The code visualizes the relationship between IMDb scores and Box Office earnings using a scatter plot. Each point represents a movie or series, with IMDb scores plotted on the x-axis and Box Office earnings on the y-axis. The plot is created using Seaborn for enhanced aesthetics and includes a grid, title, and axis labels. This visualization helps identify patterns or correlations, such as whether higher IMDb scores are associated with higher Box Office earnings.

Why scatter plot : Ideal for visualizing the relationship between two continuous variables: IMDb scores and Box Office earnings.


```python
# Question 2: What is the distribution of awards received by IMDb score?

# Plotting a scatter plot with regression line for IMDb Score vs Awards Received
plt.figure(figsize=(8,6))
sns.regplot(x='IMDb Score', y='Awards Received', data=netflix_data, scatter_kws={'color':'green'}, line_kws={'color':'red'})
plt.title('IMDb Score vs Awards Received')
plt.xlabel('IMDb Score')
plt.ylabel('Awards Received')
plt.show()
```

The code examines the relationship between IMDb scores and the number of awards received by using a scatter plot with a regression line. Each green dot represents a data point, showing the IMDb score on the x-axis and the number of awards received on the y-axis. A red regression line is included to highlight any potential linear trend. This plot helps visualize whether higher IMDb scores are associated with more awards and the overall strength of this relationship.

```python
#Question 3: Top genres by IMDb?


# Grouping by Genre and calculating the average IMDb score for each genre
top_genres_by_imdb = netflix_data.groupby('Genre')['IMDb Score'].mean().sort_values(ascending=False).head(10)

# Plotting the top 10 genres by average IMDb score using Matplotlib and Seaborn
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres_by_imdb.index, y=top_genres_by_imdb.values, palette="Blues_d")

plt.title('Top 10 Genres by Average IMDb Score', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Average IMDb Score', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

The code identifies the top 10 genres based on their average IMDb scores. It groups the dataset by genre, calculates the mean IMDb score for each genre, and sorts the results in descending order. Then, a bar chart is created using Seaborn to display the top genres, with genre names on the x-axis and average IMDb scores on the y-axis. The plot is styled with a blue color palette and includes axis labels, a rotated x-axis for better readability, and a grid for easier interpretation. This visualization highlights which genres have the highest average IMDb ratings.

Why bar chart : A bar chart is used in the above example because it is well-suited for comparing the average IMDb scores of different genres.
